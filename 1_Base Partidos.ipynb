```python
# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

# pandas -> data manipulation
import pandas as pd

# statsbombpy -> access StatsBomb Open Data
from statsbombpy import sb
```

```python
# ============================================================
# 2. LOAD ALL AVAILABLE COMPETITIONS
# ============================================================

# Load all competitions available in StatsBomb Open Data
competitions = sb.competitions()
```

```python
# ============================================================
# 3. PREVIEW COMPETITIONS
# ============================================================

# Show first rows
competitions.head()
```

```python
# ============================================================
# 4. CHECK AVAILABLE COLUMNS
# ============================================================

# Show all columns
competitions.columns
```

```python
# ============================================================
# 5. FILTER TARGET COMPETITIONS + SEASONS
# ============================================================

# Keep only the tournaments and seasons
# relevant for the project

selected_competitions = competitions[
    (
        (competitions['competition_name'] == 'FIFA World Cup') &
        (competitions['season_name'].isin(['2022']))
    )
    |
    (
        (competitions['competition_name'] == 'UEFA Euro') &
        (competitions['season_name'].isin(['2024']))
    )
    |
    (
        (competitions['competition_name'] == 'Copa America') &
        (competitions['season_name'] == '2024')
    )
    |
    (
        (competitions['competition_name'] == 'African Cup of Nations') &
        (competitions['season_name'] == '2023')
    )
]
```

```python
# ============================================================
# 6. PREVIEW FILTERED COMPETITIONS
# ============================================================

selected_competitions[
    [
        'competition_name',
        'season_name',
        'competition_id',
        'season_id'
    ]
]
```

```python
# ============================================================
# 7. CREATE MATCHES STORAGE
# ============================================================

# Empty list to store matches dataframes
matches_list = []
```

```python
# ============================================================
# 8. LOAD MATCHES FROM EVERY SELECTED COMPETITION
# ============================================================

for _, row in selected_competitions.iterrows():

    # Get competition and season IDs
    competition_id = row['competition_id']
    season_id = row['season_id']

    # Load matches from StatsBomb
    matches = sb.matches(
        competition_id=competition_id,
        season_id=season_id
    )

    # Add tournament information to each match
    matches['competition_name'] = row['competition_name']
    matches['season_name'] = row['season_name']

    # Store matches dataframe
    matches_list.append(matches)

    # Print progress
    print(f"Loaded: {row['competition_name']} - {row['season_name']}")
```

```python
# ============================================================
# 9. MERGE ALL MATCHES INTO ONE DATAFRAME
# ============================================================

matches_df = pd.concat(
    matches_list,
    ignore_index=True
)
```

```python
# ============================================================
# 10. CHECK MATCHES DATABASE SIZE
# ============================================================

print(matches_df.shape)
```

```python
# ============================================================
# 11. CHECK AVAILABLE MATCH COLUMNS
# ============================================================

matches_df.columns
```

```python
# ============================================================

# 12. PREVIEW MATCHES DATABASE

# ============================================================

matches_df.head()
```

```python
# ============================================================
# 13. KEEP IMPORTANT COLUMNS
# ============================================================

matches_df = matches_df[
    [
        'match_id',
        'match_date',
        'kick_off',

        'competition_name',
        'season_name',
        'competition_stage',

        'home_team',
        'away_team',
        'home_score',
        'away_score',

        'stadium',

        'referee'
    ]
]

# Preview selected columns
matches_df.head()
```

```python
# ============================================================
# 14. CREATE TARGET NATIONAL TEAMS LIST
# ============================================================

# List of national teams that will be used as the project universe
# Names are written in English to match StatsBomb team names

world_cup_teams = [
    'Germany',
    'Colombia',
    'Haiti',
    'Paraguay',

    'Saudi Arabia',
    'South Korea',
    'England',
    'Portugal',

    'Algeria',
    "Côte d'Ivoire",
    'Iraq',
    'Czech Republic',

    'Argentina',
    'Croatia',
    'Iran',
    'Congo DR',

    'Australia',
    'Curaçao',
    'Japan',
    'Senegal',

    'Austria',
    'Ecuador',
    'Jordan',
    'South Africa',

    'Belgium',
    'Egypt',
    'Morocco',
    'Sweden',

    'Bosnia and Herzegovina',
    'Scotland',
    'Mexico',
    'Switzerland',

    'Brazil',
    'Spain',
    'Norway',
    'Tunisia',

    'Cape Verde',
    'United States',
    'New Zealand',
    'Turkey',

    'Canada',
    'France',
    'Netherlands',
    'Uruguay',

    'Qatar',
    'Ghana',
    'Panama',
    'Uzbekistan'
]
```

```python
# ============================================================
# 15. FILTER MATCHES WITH TARGET NATIONAL TEAMS
# ============================================================

filtered_matches = matches_df[
    (matches_df['home_team'].isin(world_cup_teams)) |
    (matches_df['away_team'].isin(world_cup_teams))
].reset_index(drop=True)
```

```python
# ============================================================
# 16. PREVIEW FILTERED MATCHES
# ============================================================

filtered_matches.sample(5)
```

```python
# ============================================================
# 17. TOTAL FILTERED MATCHES
# ============================================================

print(f"Total matches: {len(filtered_matches)}")
```

```python
# ============================================================
# 18. MATCH COUNT BY TEAM
# ============================================================

home_counts = filtered_matches['home_team'].value_counts()
away_counts = filtered_matches['away_team'].value_counts()

team_match_counts = home_counts.add(
    away_counts,
    fill_value=0
).sort_values(ascending=False)

team_match_counts
```

```python
# ============================================================
# SET FINAL MATCHES BASE
# ============================================================

matches = filtered_matches.copy()

print(matches.shape)
print(matches["competition_name"].value_counts())
```

```python
# ============================================================
# CLEAN MATCHES BASE
# ============================================================

matches["match_date"] = pd.to_datetime(matches["match_date"])
matches["season_name"] = matches["season_name"].astype(str)

matches["match_label"] = (
    matches["match_date"].dt.strftime("%Y-%m-%d") + " | " +
    matches["home_team"] + " " +
    matches["home_score"].astype(str) + "-" +
    matches["away_score"].astype(str) + " " +
    matches["away_team"]
)

def get_result(row, side):
    if row["home_score"] == row["away_score"]:
        return "Draw"

    if side == "home":
        return "Win" if row["home_score"] > row["away_score"] else "Loss"

    if side == "away":
        return "Win" if row["away_score"] > row["home_score"] else "Loss"

matches["home_result"] = matches.apply(lambda row: get_result(row, "home"), axis=1)
matches["away_result"] = matches.apply(lambda row: get_result(row, "away"), axis=1)
```

```python
# ============================================================
# COMPETITION SHORT NAME + TOURNAMENT ID
# ============================================================

competition_mapping = {
    "FIFA World Cup": "World Cup",
    "UEFA Euro": "Euro",
    "Copa America": "Copa América",
    "African Cup of Nations": "AFCON"
}

matches["competition_short"] = matches["competition_name"].map(competition_mapping)

matches["tournament_id"] = (
    matches["competition_short"] + "_" +
    matches["season_name"].astype(str)
)

matches.head()
```

```python
# ============================================================
# SAVE MATCHES BASE
# ============================================================

matches.to_csv("matches.csv", index=False)

print("matches.csv saved successfully!")
print(matches["competition_name"].value_counts())
```
