```python
# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
from statsbombpy import sb
```

```python
# ============================================================
# 2. LOAD MATCHES BASE
# ============================================================

matches = pd.read_csv("matches.csv")

print("Matches:", matches.shape)
matches.head()
```

```python
# ============================================================
# 3. LOAD EVENTS FOR ALL MATCHES
# ============================================================

all_events_list = []

for match_id in matches["match_id"].unique():

    try:

        events = sb.events(match_id=match_id)

        events["match_id"] = match_id

        match_info = matches[
            matches["match_id"] == match_id
        ].iloc[0]

        events["competition_name"] = match_info["competition_name"]

        events["season_name"] = str(
            match_info["season_name"]
        )

        events["competition_short"] = (
            match_info["competition_short"]
        )

        events["tournament_id"] = (
            match_info["tournament_id"]
        )

        all_events_list.append(events)

        print(f"✓ Loaded events for match {match_id}")

    except Exception as e:

        print(f"✗ Error loading match {match_id}: {e}")

events = pd.concat(
    all_events_list,
    ignore_index=True
)

print("Events:", events.shape)

events.head()
```

```python
# ============================================================
# 4. CREATE TEAM MATCH STATS
# ============================================================

team_match_stats = (
    events
    .groupby([
        "competition_name",
        "season_name",
        "competition_short",
        "tournament_id",
        "match_id",
        "team"
    ])
    .agg(
        total_events=("type", "count"),

        shots=("type",
                lambda x: (x == "Shot").sum()),

        passes=("type",
                 lambda x: (x == "Pass").sum()),

        carries=("type",
                  lambda x: (x == "Carry").sum()),

        pressures=("type",
                    lambda x: (x == "Pressure").sum()),

        recoveries=("type",
                     lambda x: (x == "Ball Recovery").sum()),

        duels=("type",
                lambda x: (x == "Duel").sum()),

        interceptions=("type",
                         lambda x: (x == "Interception").sum()),

        blocks=("type",
                  lambda x: (x == "Block").sum()),

        clearances=("type",
                      lambda x: (x == "Clearance").sum())
    )
    .reset_index()
)

team_match_stats.head()
```

```python
# ============================================================
# 5. ADD TEAM XG
# ============================================================

team_xg = (
    events[
        events["type"] == "Shot"
    ]
    .groupby([
        "match_id",
        "team"
    ])
    .agg(
        xg=("shot_statsbomb_xg", "sum")
    )
    .reset_index()
)

team_match_stats = team_match_stats.merge(
    team_xg,
    on=[
        "match_id",
        "team"
    ],
    how="left"
)

team_match_stats["xg"] = (
    team_match_stats["xg"]
    .fillna(0)
)

team_match_stats.head()
```

```python
# ============================================================
# 6. ADD TEAM COMPLETED PASSES
# ============================================================

completed_passes_team = (
    events[
        (events["type"] == "Pass") &
        (events["pass_outcome"].isna())
    ]
    .groupby([
        "match_id",
        "team"
    ])
    .size()
    .reset_index(
        name="completed_passes"
    )
)

team_match_stats = team_match_stats.merge(
    completed_passes_team,
    on=[
        "match_id",
        "team"
    ],
    how="left"
)

team_match_stats["completed_passes"] = (
    team_match_stats["completed_passes"]
    .fillna(0)
    .astype(int)
)

team_match_stats["pass_completion_pct"] = np.where(
    team_match_stats["passes"] > 0,

    (
        team_match_stats["completed_passes"] /
        team_match_stats["passes"] * 100
    ),

    0
)

team_match_stats.head()
```

```python
# ============================================================
# 7. CREATE PLAYER MATCH STATS
# ============================================================

player_match_stats = (
    events[
        events["player"].notna()
    ]
    .groupby([
        "competition_name",
        "season_name",
        "competition_short",
        "tournament_id",
        "match_id",
        "team",
        "player"
    ])
    .agg(
        total_events=("type", "count"),

        shots=("type",
                lambda x: (x == "Shot").sum()),

        passes=("type",
                 lambda x: (x == "Pass").sum()),

        carries=("type",
                  lambda x: (x == "Carry").sum()),

        pressures=("type",
                    lambda x: (x == "Pressure").sum()),

        recoveries=("type",
                     lambda x: (x == "Ball Recovery").sum()),

        duels=("type",
                lambda x: (x == "Duel").sum()),

        interceptions=("type",
                         lambda x: (x == "Interception").sum()),

        blocks=("type",
                  lambda x: (x == "Block").sum()),

        clearances=("type",
                      lambda x: (x == "Clearance").sum())
    )
    .reset_index()
)

player_match_stats.head()
```

```python
# ============================================================
# 8. ADD PLAYER XG
# ============================================================

player_xg = (
    events[
        (events["type"] == "Shot") &
        (events["player"].notna())
    ]
    .groupby([
        "match_id",
        "team",
        "player"
    ])
    .agg(
        xg=("shot_statsbomb_xg", "sum")
    )
    .reset_index()
)

player_match_stats = player_match_stats.merge(
    player_xg,
    on=[
        "match_id",
        "team",
        "player"
    ],
    how="left"
)

player_match_stats["xg"] = (
    player_match_stats["xg"]
    .fillna(0)
)

player_match_stats.head()
```

```python
# ============================================================
# 9. ADD PLAYER COMPLETED PASSES
# ============================================================

completed_passes_player = (
    events[
        (events["type"] == "Pass") &
        (events["pass_outcome"].isna()) &
        (events["player"].notna())
    ]
    .groupby([
        "match_id",
        "team",
        "player"
    ])
    .size()
    .reset_index(
        name="completed_passes"
    )
)

player_match_stats = player_match_stats.merge(
    completed_passes_player,
    on=[
        "match_id",
        "team",
        "player"
    ],
    how="left"
)

player_match_stats["completed_passes"] = (
    player_match_stats["completed_passes"]
    .fillna(0)
    .astype(int)
)

player_match_stats["pass_completion_pct"] = np.where(
    player_match_stats["passes"] > 0,

    (
        player_match_stats["completed_passes"] /
        player_match_stats["passes"] * 100
    ),

    0
)

player_match_stats.head()
```

```python
# ============================================================
# 10. CREATE PLAYER POSITIONS BASE
# ============================================================

positions_list = []

for match_id in matches["match_id"].unique():

    try:
        lineups = sb.lineups(match_id=match_id)

        match_info = matches[
            matches["match_id"] == match_id
        ].iloc[0]

        for team_name, lineup_df in lineups.items():

            for _, player_row in lineup_df.iterrows():

                player_name = player_row["player_name"]
                jersey_number = player_row.get("jersey_number", np.nan)
                positions_data = player_row.get("positions", [])

                if isinstance(positions_data, list) and len(positions_data) > 0:

                    first_position = positions_data[0]

                    position_name = first_position.get("position", np.nan)
                    position_id = first_position.get("position_id", np.nan)
                    position_from = first_position.get("from", np.nan)
                    position_to = first_position.get("to", np.nan)

                    starter = position_from == "00:00"

                else:

                    position_name = np.nan
                    position_id = np.nan
                    position_from = np.nan
                    position_to = np.nan
                    starter = False

                positions_list.append({
                    "competition_name": match_info["competition_name"],
                    "season_name": str(match_info["season_name"]),
                    "competition_short": match_info["competition_short"],
                    "tournament_id": match_info["tournament_id"],
                    "match_id": match_id,
                    "team": team_name,
                    "player": player_name,
                    "position": position_name,
                    "position_id": position_id,
                    "from": position_from,
                    "to": position_to,
                    "jersey_number": jersey_number,
                    "starter": starter
                })

        print(f"✓ Loaded positions for match {match_id}")

    except Exception as e:

        print(f"✗ Error loading positions for match {match_id}: {e}")

positions = pd.DataFrame(positions_list)

print("Positions:", positions.shape)

positions.head()
```

```python
# ============================================================
# 11. ADD POSITION COORDINATES
# ============================================================

position_coordinates = {
    "Goalkeeper": (8, 40),

    "Left Back": (28, 12),
    "Left Center Back": (25, 28),
    "Center Back": (25, 40),
    "Right Center Back": (25, 52),
    "Right Back": (28, 68),

    "Left Wing Back": (38, 10),
    "Right Wing Back": (38, 70),

    "Left Defensive Midfield": (45, 28),
    "Center Defensive Midfield": (45, 40),
    "Right Defensive Midfield": (45, 52),

    "Left Center Midfield": (60, 28),
    "Center Midfield": (60, 40),
    "Right Center Midfield": (60, 52),

    "Left Midfield": (62, 14),
    "Right Midfield": (62, 66),

    "Left Attacking Midfield": (78, 28),
    "Center Attacking Midfield": (78, 40),
    "Right Attacking Midfield": (78, 52),

    "Left Wing": (88, 12),
    "Right Wing": (88, 68),

    "Left Center Forward": (100, 32),
    "Center Forward": (105, 40),
    "Right Center Forward": (100, 48),

    "Secondary Striker": (92, 40)
}

position_xy = pd.DataFrame([
    {
        "position": position,
        "position_x": coords[0],
        "position_y": coords[1]
    }
    for position, coords in position_coordinates.items()
])

positions = positions.merge(
    position_xy,
    on="position",
    how="left"
)

positions.head()
```

```python
# ============================================================
# 12. CHECK POSITIONS WITHOUT COORDINATES
# ============================================================

positions_without_xy = (
    positions[
        positions["position_x"].isna()
    ]["position"]
    .dropna()
    .unique()
)

positions_without_xy
```

```python
# ============================================================
# 13. MERGE PLAYER STATS WITH POSITIONS
# ============================================================

player_match_stats_with_positions = player_match_stats.merge(
    positions[
        [
            "match_id",
            "team",
            "player",
            "position",
            "position_id",
            "position_x",
            "position_y",
            "jersey_number",
            "starter",
            "from",
            "to"
        ]
    ],
    on=["match_id", "team", "player"],
    how="left"
)

player_match_stats_with_positions.head()
```

```python
# ============================================================
# 14. SAVE OUTPUTS
# ============================================================

events.to_csv("events.csv", index=False)

team_match_stats.to_csv(
    "team_match_stats.csv",
    index=False
)

player_match_stats.to_csv(
    "player_match_stats.csv",
    index=False
)

positions.to_csv(
    "positions.csv",
    index=False
)

player_match_stats_with_positions.to_csv(
    "player_match_stats_with_positions.csv",
    index=False
)

print("Saved:")
print("events.csv")
print("team_match_stats.csv")
print("player_match_stats.csv")
print("positions.csv")
print("player_match_stats_with_positions.csv")
```
