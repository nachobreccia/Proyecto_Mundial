```python
import pandas as pd
from pathlib import Path

events_path = Path("events.csv")
data_dir = Path("streamlit_app/data")
data_dir.mkdir(parents=True, exist_ok=True)

# borrar partes anteriores
for old_file in data_dir.glob("events_part_*.csv"):
    old_file.unlink()

df = pd.read_csv(events_path)

chunk_size = 100000

for i, start in enumerate(range(0, len(df), chunk_size), start=1):
    out = data_dir / f"events_part_{i}.csv"
    df.iloc[start:start + chunk_size].to_csv(out, index=False)
    print(out.name, round(out.stat().st_size / 1024 / 1024, 1), "MB")
```
