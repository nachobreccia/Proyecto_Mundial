```python
# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================

import os
import shutil
from pathlib import Path
```

```python
# ============================================================
# 2. CREATE STREAMLIT DATA FOLDER
# ============================================================

streamlit_data_path = Path(
    "streamlit_app/data"
)

streamlit_data_path.mkdir(
    parents=True,
    exist_ok=True
)

print("streamlit_app/data ready!")
```

```python
# ============================================================
# 3. FINAL CSV FILES
# ============================================================

final_csvs = [
    "matches.csv",
    "events.csv",
    "team_match_stats.csv",
    "player_match_stats.csv",
    "positions.csv",
    "player_match_stats_with_positions.csv",
    "team_master.csv",
    "player_master.csv",
    "team_similarity.csv",
    "player_similarity.csv",
    "highlighted_matches.csv",
    "most_frequent_xi.csv",
    "player_role_fit.csv",
    "assets.csv"
]

print(f"Files configured: {len(final_csvs)}")
print(f"Unique files: {len(set(final_csvs))}")

if len(final_csvs) != len(set(final_csvs)):
    duplicated_files = [
        file for file in final_csvs
        if final_csvs.count(file) > 1
    ]

    raise ValueError(
        f"Duplicated files detected: {set(duplicated_files)}"
    )
```

```python
# ============================================================
# 4. COPY FILES TO STREAMLIT
# ============================================================

copied_files = []
missing_files = []

for file_name in final_csvs:

    if os.path.exists(file_name):

        shutil.copy2(
            file_name,
            streamlit_data_path / file_name
        )

        copied_files.append(file_name)

        file_size_mb = os.path.getsize(file_name) / (1024 * 1024)

        print(
            f"✓ Copied: {file_name} | {file_size_mb:.2f} MB"
        )

    else:

        missing_files.append(file_name)

        print(f"✗ Missing: {file_name}")
```

```python
# ============================================================
# 4B. CHECK CRITICAL FILES
# ============================================================

critical_files = [
    "matches.csv",
    "events.csv",
    "team_master.csv",
    "player_master.csv",
    "team_similarity.csv",
    "player_similarity.csv",
    "highlighted_matches.csv",
    "most_frequent_xi.csv",
    "player_role_fit.csv",
    "assets.csv"
]

missing_critical = [
    file for file in critical_files
    if file in missing_files
]

if len(missing_critical) > 0:
    raise FileNotFoundError(
        f"Critical files missing: {missing_critical}"
    )

print("✓ All critical files available")
```

```python
# ============================================================
# 5. SUMMARY
# ============================================================

print("\n====================")
print("COPY SUMMARY")
print("====================")

print(f"\nCopied files: {len(copied_files)}")

for file in copied_files:
    print(f"✓ {file}")

print(f"\nMissing files: {len(missing_files)}")

for file in missing_files:
    print(f"✗ {file}")

print("\nNotebook 8 export completed successfully!")
```
